```python
# Clean up missing data
df_2017 = dataframe_2017[dataframe_2017["Rating (Num)"] != "#ERROR!"]
df_2017["Rating (Num)"] = pd.to_numeric(df_2017["Rating (Num)"])
```

```python
# Basic Information on 2017
import seaborn as sb

# Sort by Week
df_2017 = df_2017.sort_index(by=['Week'], ascending=[True])

# How many surveys we have in each week?
sb.countplot(x="Week", data=df_2017).set_title("Survey Submissions")
```

```python
# Weekly NPS (2017)
def calcNPS(df):
    promoter  = df[df["Rating (Num)"] > 8]
    detractor = df[df["Rating (Num)"] < 7]
    nps = (promoter["Rating (Num)"].count() - detractor["Rating (Num)"].count()) / df["Rating (Num)"].count()
    return nps

df_weeks = df_2017.groupby("Week")
df_weeks.apply(lambda x: calcNPS(x)).plot(title="NPS (2017)")
```

Seems good! How about each location?

```python
# Location
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15,7))
df_week_loc = df_2017.groupby(["Week","Location"])
df_week_loc.apply(lambda x: calcNPS(x)).unstack().plot(ax=ax, title="NPS Each Location")
```
