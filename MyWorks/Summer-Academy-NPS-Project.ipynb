```python
# Clean up missing data
df_2017 = dataframe_2017[dataframe_2017["Rating (Num)"] != "#ERROR!"]
```

```python
dataframe_2017.describe()
```

```python
df_2017["Rating (Num)"] = pd.to_numeric(df_2017["Rating (Num)"])
```

```python
# Basic Information on 2017
import seaborn as sb

# Sort by Week
df_2017 = df_2017.sort_index(by=['Week'], ascending=[True])

# How many surveys we have in each week?
sb.countplot(x="Week", data=df_2017)
```

```python
# How many surveys we have in each week?
sb.countplot(x="Week", data=df_2017)
```

```python
# Weekly NPS (2017)
def calcNPS(df):
    promoter  = df[df["Rating (Num)"] > 8]
    detractor = df[df["Rating (Num)"] < 7]
    nps = (promoter["ID"].count() - detractor["ID"].count()) / df["ID"].count()
    return nps

df_weeks = df_2017.groupby("Week")
df_weeks.apply(lambda x: calcNPS(x)).plot()
```

```python
# Location
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15,7))
df_week_loc = df_2017.groupby(["Week","Location"])
df_week_loc.apply(lambda x: calcNPS(x)).unstack().plot(ax=ax)
```

```python
# Track
fig, ax = plt.subplots(figsize=(15,7))
df_track = df_2017.groupby(["Week","Track"])
df_track.apply(lambda x: calcNPS(x)).unstack().plot(ax=ax)
```

```python
# Pacing
fig, ax = plt.subplots(figsize=(15,7))
df_track = df_2017.groupby(["Week","Schedule Pacing"])
df_track.apply(lambda x: calcNPS(x)).unstack().plot(ax=ax)
```
